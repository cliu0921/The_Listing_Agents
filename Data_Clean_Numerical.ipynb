```python
dict_ = {}
for i in range(len(homes_num_columns)):

    col_ = homes_num_columns[i]
    col_val_ = homes[col_]
    variance_ = np.var(col_val_,0)
    lis_ = []
    lis_.append(variance_)
    dict_[col_] = lis_
```
